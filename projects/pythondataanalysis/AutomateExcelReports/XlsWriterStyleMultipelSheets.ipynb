```python
import pandas as pd
from xlsxwriter.utility import xl_rowcol_to_cell
```

```python
df = pd.read_excel("AdvWorksData.xlsx")
df.head()
```

```python
df1 = df.iloc[1:10]
df2 = df.iloc[11:20]
```

```python
df1 = df1[['productcategory','productsubcategory','product','Country','OrderDate','StandardCost','Discount','Sales']]
df2 = df2[['productcategory','productsubcategory','product','Country','OrderDate','StandardCost','Discount','Sales']]
```

```python
df2
```

```python
writer = pd.ExcelWriter('multisheets.xlsx', engine='xlsxwriter')
sheet_list=['Sheet1','sheet2']
df_list=[df1, df2]

for i,j in zip(df_list,sheet_list):
    i.to_excel(writer,j,index=False)

### Assign WorkBook
workbook=writer.book

#Set header formating
header_format = workbook.add_format({
        "valign": "vcenter",
        "align": "center",
        "bg_color": "#951F06",
         "bold": True,
        'font_color': '#FFFFFF'
    })
### Apply same format on each sheet being saved
for i,j in zip(df_list,sheet_list):
    #add title
    title = "Monthly Sales Report "
    #merge cells
    format = workbook.add_format()
    format.set_font_size(20)
    format.set_font_color("#333333")
    #
    subheader = "Sales report for Classic Vest, M"
    writer.sheets[j].merge_range('A1:AS1', title, format)
    writer.sheets[j].merge_range('A2:AS2', subheader)
    writer.sheets[j].set_row(2, 15) # Set the header row height to 15
    for col_num, value in enumerate(i.columns.values):
        writer.sheets[j].write(2, col_num, value, header_format)
        # Adjust the column width.
        writer.sheets[j].set_column('A:F', 20)
writer.save()
```
