```python
import pandas as pd
```

```python
popgrowth = pd.read_csv('data/population-growth.csv')
popgrowth
```

```python
worldgrowth = popgrowth[popgrowth['Country name'] == 'World']
worldgrowth
```

```python
popdf = pd.read_csv('data/population.csv')
worldpop = popdf[popdf['Country name'] == 'World']
worldpop
```

### World population projection

```python
worldpop.plot(x='Year',y='Population')
```

### India will overtake China as most the populous country

```python
ax = popdf[popdf['Country name'] == 'India'].plot(x = 'Year',y = 'Population', label = 'India')
popdf[popdf['Country name'] == 'China'].plot(x = 'Year', label = "China", y='Population', ax = ax)
```
